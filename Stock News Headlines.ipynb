```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']
```

```python
train.head()
```

```python
data=train.iloc[:,2:27]
data.head()
```

```python
data.columns
```

```python
data.replace('[^a-zA-Z]',' ',inplace=True)
```

```python
data.head()
```

```python
data.columns
```

```python
new_index=[str(i) for i in range(25)]
new_index
```

```python
data.columns=new_index
data.head()
```

```python
data.index
```

```python
for index in new_index:
    data[index]=data[index].str.lower()
data.head()   
```

```python
#Data Wrangling and Merging
```

```python
data.iloc[1,0:25]
```

```python
headlines=[]
for i in data.iloc[1,0:25]:
    headlines.append(i)
' '.join(headlines )   
```

```python
' '.join([str(i) for i in data.iloc[1,0:25]])
```

```python
headlines=[]
for row in range(0,len(data)):
    headlines.append(' '.join([str(i) for i in data.iloc[row,0:25]]))
```

```python
headlines[0:3]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv=CountVectorizer(ngram_range=(2,2))
traindata_x=cv.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
```

```python
randomclassifier.fit(traindata_x,train['Label'])
```

```python
test.head()
```

```python
test_transform=[]
for row in range(0,len(test)):
    test_transform.append(' '.join([str(i) for i in test.iloc[row,2:27]]))
```

```python
test_data=cv.transform(test_transform)
```

```python
predictions=randomclassifier.predict(test_data)
predictions
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(test['Label'],predictions)
cm
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.imshow(cm,cmap=plt.cm.Blues)
plt.title('confusion matrix')
plt.colorbar()
labels=['positive','negative']
tick_marks=np.arange(len(labels))
plt.xticks(tick_marks,labels,rotation=45)
plt.yticks(tick_marks,labels)
plt.tight_layout()
plt.xlabel('true label')
plt.ylabel('predicted labels')
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
accuracy_score(test['Label'],predictions)
```

```python
report=classification_report(test['Label'],predictions)
print(report)
```

```python
# BAYE'S THEOREM
#p(a/b)=(p(b/a)*p(a))/p(b) 
#p(b)=marginal probability
#p(a)=prior probability
#p(b/a)=likelihood
#p(a/b)=posterior probability
```

```python
def plot_confusion_matrix(cm,title='Confusion matrix'):
    plt.imshow(cm,cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar()
    labels=['positive','negative']
    tick_marks=np.arange(len(labels))
    plt.xticks(tick_marks,labels,rotation=45)
    plt.yticks(tick_marks,labels)
    plt.tight_layout()
    plt.xlabel('true label')
    plt.ylabel('predicted labels')
```

```python
plot_confusion_matrix(cm2)
```
